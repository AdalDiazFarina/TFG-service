# Downloading data

```python
def download_data(ticket):
        # Descargar datos de yfinance
        data = yf.download(ticket, start="2011-01-01", end="2024-01-01", interval="1mo")
        return data
```

```python
weights = np.array([0.25, 0.25, 0.25, 0.25])
monthly_contribution = 100
initial_capital = 1000
target_weights = weights.copy()
values = initial_capital * weights 
```

# Rebalance


```python
def rebalance():
    total_value = np.sum(values)  # Calcular el valor total de la cartera
    target_value = total_value * target_weights  # Calcular el valor objetivo para cada activo
    for i in range(len(values)):
        # Verificar si algÃºn activo se desvÃ­a mÃ¡s del 5% de su objetivo
        if values[i] > target_value[i] * 1.05 or values[i] < target_value[i] * 0.95:
            values = total_value * target_weights  # Rebalancear los activos
            break
```

# Monthly update

```python
def monthly_update(price_changes):
        # Actualizar los valores de los activos con los cambios porcentuales de precios
        values *= (1 + price_changes)
        # AÃ±adir las aportaciones mensuales
        values += monthly_contribution * target_weights
        # Rebalancear la cartera si es necesario
        rebalance()
```
