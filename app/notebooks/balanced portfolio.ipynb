```python
weights = np.array([0.25, 0.25, 0.25, 0.25])
monthly_contribution = 100
initial_capital = 1000
values = initial_capital * weights 
operations = [pd.DataFrame(columns=['Asset', 'Date', 'Type', 'Amount', 'Price', 'Total','Period']) for _ in range(4)]
```

# Rebalance


```python
def Redistribution(redistribution_flags, redistributed_values, timelines, row, period):
  for i in range(len(values)):
        if redistribution_flags[i]:  # If the value has fallen more than 8%
            # We remove 15% of the asset value
            redistribution_amount = values[i] * 0.15
            redistributed_values[i] -= redistribution_amount
            operations[i].loc[len(operations[i])] = {
                'Asset': 'SP&500',
                'Date': timelines[i][row]['Date'],
                'Type': 'Sell',
                'Amount': redistribution_amount,
                'Price': timelines[i][row]['Open'],
                'Total': redistributed_values[i],
                'Period': period
            }
            
            # Redistribute among the other assets that have not fallen more than 8%
            non_fallen_indices = [j for j in range(len(values)) if not redistribution_flags[j]]
            if non_fallen_indices:
                redistribution_per_asset = redistribution_amount / len(non_fallen_indices)
                for j in non_fallen_indices:
                    redistributed_values[j] += redistribution_per_asset   
    
  return redistributed_values
```

```python
def MonthlyUpdate(movements, timelines, row, period):
    # Updating the current amount of each asset
    values += values * movements
    # Selecting the asset than has fallen more than 8%
    redistribution_flags = [movements[i] > 0.08 for i in range(len(movements))]
    
    # No redistribution if all values have fallen more than 8%
    if all(redistribution_flags):
        return values
    
    values = Redistribution(values, redistribution_flags, timelines, row, period)
    # Monthly contribution
    contributions = monthly_contribution * weights
    values +=  contributions - (contributions * comission / 100)

    
```

```python
buySell_comission = 0.12
current_exchange_rate = 0.30
comission = buySell_comission + current_exchange_rate
```

# Simulate

```python
## This method calculate the total return of one active
def TotalReturn(timelines, period, initial_capital=initial_capital):
  if (initial_capital is None): initial_capital = initial_capital
  for i in range(len(timelines[0])):
    movements = []
    for j in range(len(timelines)):
      ## Percentage of asset movement
      movements.push((timelines[j][i]['Close'] - timelines[j][i]['Open']) / timelines[j][i]['Open'])
      MonthlyUpdate(movements, operations, timelines, i, period)
      ## Ading the dividend
      if not pd.isna(timelines[j][i]['Dividends']):
        values[j] += values[j] * timelines[j][i]['Dividends'] / 100

  return {
    'total': values,
    'operations': operations
  }
```
